```python
#1+
import random
a=[random.randint(1, 10) for i in range(10)]
print(a)
n=int(input('Введите число: '))
def f(a,n):
    b=[]
    for i in a:
        if i%n==0:
            b.append(i)
    return b
print(f(a,n))
```

```python
#2
import random
a=[random.randint(1, 10) for i in range(10)]
print(a)
def f(a):
    b=[]
    for i in range(len(a)):
        if (i+1) %5 == 0:
            b.append(a[i]*2)
        if (i+1) %3 != 0:
            b.append(a[i])        
    return b
print(f(a))
```

```python
#3+
import random
a=[random.randint(1, 10) for i in range(10)]
print(a)
n=int(input('Введите число: '))
def f(a,n):
    sum=0
    for i in a:
        if i==n:
            sum+=1
    return sum
print(f(a,n))
```

```python
#4+
import random
a=[random.randint(1, 10) for i in range(9)]
print(a)
def f(c):
    b=[]
    i = 0
    for j in range(int(len(c)/3)):
        a=c[i:i+3]
        b.extend(a[::-1])
        i+=len(a)
    return b
print(f(a))
```

```python
#5+
import random
a=[random.randint(1, 10) for i in range(10)]
b=[random.randint(1, 10) for i in range(10)]
print(a)
print(b)
def f(a,b):
    c=[]
    for i in range(len(a)):
        c.append(a[i]-b[i])
    return c
print(f(a,b))
```

```python
#6+
import random
a = [random.randint(1, 10) for i in range(10)]
print(a)
def f(a):
    b=[]
    works=[]
    for i in a:
        if i in works:
            continue
        b.append(a.count(i))
        works.append(i)
    return b
print(f(a))
```

```python
data_male = data[data['Пол']=='м']
data_female = data[data['Пол']=='ж']
display(data_male)
display(data_female)
```

```python
import datetime as dt

dataM = data[data['Пол']=='м']
dataM['Дата приема на работу'] = pd.to_datetime(dataM['Дата приема на работу'])

dataM.loc[:, 'Стаж'] = (np.floor((pd.to_datetime(dt.date.today().strftime('%Y-%m-%d')) - dataM['Дата приема на работу']).dt.days / 365.25)).astype(int)
dataM
```
